```python
#relationship analysis
#correlation matrix

corelation = stud.corr()
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
sns.heatmap(corelation, xticklabels= corelation.columns, yticklabels= corelation.columns,
            annot = True)
```

```python
#relationship between two categorical, continous or boolean variable
sns.pairplot(stud)
```

```python
#scatter plot: relation between two numerical values
sns.relplot(x ='math_score', y ='reading_score', hue='gender', data= stud)
```

```python
#display of graph using different heights: histogram
sns.displot(stud['reading_score'])
```

```python
#categorical box plot
sns.catplot(x ='math_score', kind = 'box', data = stud)
```
